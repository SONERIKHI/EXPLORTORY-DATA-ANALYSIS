```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
tip = pd.read_csv("tips.csv")
tip
```

```python
tips = tip.drop(["price_per_person","Payer Name","CC Number","Payment ID"],axis = 1)
```

```python
tips
```

```python
tips.info()
```

```python
tips.isnull()
```

```python
tips.isnull().sum()
```

```python
plt.figure(figsize=(18,5))
plt.grid("tips",linestyle=":")
plt.hist(tips["total_bill"],bins=20,color="maroon",edgecolor="yellow")
```

```python
sns.relplot(x ='total_bill', y = 'tip', data = tips,height=4,aspect=3)
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.grid("tips",linestyle=":")
plt.hist(tips["sex"],bins=20,color="maroon",edgecolor="yellow")
```

```python
sns.relplot(x = 'total_bill', y = "tip", data = tips,hue="sex",height=5,aspect=3)
```

```python
sns.relplot(x = 'total_bill', y = "tip", data = tips,col="sex")
```

```python
plt.figure(figsize=(18,5))
plt.grid("tips",linestyle=":")
plt.hist(tips["smoker"],bins=20,color="maroon",edgecolor="yellow")
```

```python
sns.relplot(x = 'total_bill', y = "smoker", data = tips,hue="sex")
```

```python
sns.relplot(x = 'total_bill', y = "smoker", data = tips,col="sex")
```

```python
sns.relplot(x = 'total_bill', y = "tip", data = tips,hue="size")
```

```python
sns.relplot(x = 'total_bill', y = "tip", data = tips,hue="size", col= "sex")
```

```python
sns.relplot(x = 'total_bill', y = "tip", data = tips,hue="sex",col="size")
```

```python
sns.catplot(x = 'total_bill', y = "time", data = tips,kind="box",height=5,aspect=2.5)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,aspect=2.5,hue="time",kind="boxen")
```
